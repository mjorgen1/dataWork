# FULL ALTPROV DATA 

```python
altprov_data = pd.read_csv('spc_alternative_provision_2021.csv')
```

```python
# with 77722 rows × 15 columns
altprov_data
```

## DELETE Unecessary Columns...

```python
altprov_data = altprov_data.drop(labels=['time_identifier','country_name','country_code','region_code','old_la_code','new_la_code'], axis=1)
print(altprov_data)
```

```python
# To check for uniqueness of values in columns...
#print(altprov_data.geographic_level.unique())
```

## DELETE nonzero pupil number rows AND Not FSM eligible rows

```python
# How to select rows with multiple conditions: df.loc[(df['col1'] == value) & (df['col2'] < value)]
nonzero_altprov_data = altprov_data.loc[(altprov_data['number_of_pupils'] != 0) & (altprov_data['FSM'] != 'Not FSM eligible')]
```

```python
# NOW WE HAVE
# 33118 rows × 9 columns of FSM Eligible and Number of Pupils being 0
print(nonzero_altprov_data)
```

## SPLIT DATA by Geographic Level

```python
# Local Auth level with 29786 rows
la_altprov_data = nonzero_altprov_data.loc[nonzero_altprov_data['geographic_level'] == 'Local authority']
la_altprov_data = la_altprov_data.drop(labels=['geographic_level'],axis=1)
# Regional Level with 2940 rows
regional_altprov_data = nonzero_altprov_data.loc[nonzero_altprov_data['geographic_level'] == 'Regional']
regional_altprov_data = regional_altprov_data.drop(labels=['geographic_level','la_name'],axis=1)
# National Level with 392 rows
national_altprov_data = nonzero_altprov_data.loc[nonzero_altprov_data['geographic_level'] == 'National']
national_altprov_data = national_altprov_data.drop(labels=['geographic_level','region_name','la_name'],axis=1)
```
